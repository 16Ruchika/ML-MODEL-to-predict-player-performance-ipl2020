```python
df_train1= pd.read_csv("../input/ipl-2020-player-performance/Training.csv")
```

```python
X = np.array(df_train1.iloc[:,1:-1])
y = np.array(df_train1.iloc[:,-1])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 , random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train.shape
```

```python
X_test = scaler.transform(X_test)
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics

regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
regression_model.score(X_test, y_test)
```

```python
predicted_train_lr=regression_model.fit(X_train, y_train).predict(X_train)
np.sqrt(metrics.mean_squared_error(y_train,predicted_train_lr))
```

```python
from math import sqrt 
from sklearn.metrics import mean_squared_error
predicted_test_lr = regression_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test,predicted_test_lr))
rmse
```

```python
predicted_test_lr = np.round(predicted_test_lr)
predicted_test_lr
```
